```python
import pandas as pd
import shutil
import librosa
import numpy as np
from scipy.io import wavfile
```

```python
basePath = "/home/birdo/MachineLearning/"
sourcePath = basePath+"Data/esc50/audio/"
destPath = basePath+"Data/esc50/audio_no_bird"
```

```python
df = pd.read_csv(basePath + "Data/esc50/esc50.csv")
```

```python
df
```

```python
df.category.unique()
```

```python
birds = df.loc[df['category'].isin(['chirping_birds', 'hen', 'rooster'])]
```

```python
birds
```

```python
no_birds = df.loc[~df['category'].isin(['chirping_birds', 'hen', 'rooster'])]
```

```python
no_birds
```

```python
no_birds_files = no_birds.filename.values
```

```python
for i in range(0, len(no_birds_files), 2):
    if(i+1 < len(no_birds_files)):
        file1, sr1 = librosa.load(sourcePath+'/'+no_birds_files[i], sr=44100)
        file2, sr2 = librosa.load(sourcePath+'/'+no_birds_files[i+1], sr=44100)
        outfile = np.append(file1, file2)
        filename = "esc50-"+str(i)+".wav"
        wavfile.write(data=outfile, rate=44100, filename = destPath+'/'+filename)
```
